```python
import tensorflow as tf
import matplotlib.pyplot as plt

X = [1, 2, 3]
Y = [1, 2, 3]

W = tf.placeholder(tf.float32)
hypothesis = X * W

cost = tf.reduce_mean(tf.square(hypothesis - Y)) #hypothesis와 y의 차를 제곱한다

W_history = []
cost_history = [] #그래프를 그리기 위한 것, 값을 저장

with tf.Session() as sess:
    for i in range(-30, 50):
        curr_W = i * 0.1
        curr_cost = sess.run(cost, feed_dict={W: curr_W})

        W_history.append(curr_W) #리스트에 넣음
        cost_history.append(curr_cost)

plt.plot(W_history, cost_history) #그래프를 그림
plt.show()
```

```python
import tensorflow as tf

x_data = [1, 2, 3]
y_data = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name="weight")

X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32) #값을 넘겨줌

hypothesis = X * W
cost = tf.reduce_mean(tf.square(hypothesis - Y))

learning_rate = 0.1 
gradient = tf.reduce_mean((W * X - Y) * X) #수동으로 cost를 minimize함
descent = W - learning_rate * gradient #수동 gradient decent 해봄
update = W.assign(descent)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(21):
        sess.run(update, feed_dict = {X: x_data, Y: y_data})
        print(step, sess.run(cost, feed_dict = {X: x_data, Y: y_data}), sess.run(W))
```

```python
import tensorflow as tf

X = [1, 2, 3]
Y = [1, 2, 3]

W = tf.Variable(-3.0)

hypothesis = X * W
cost = tf.reduce_mean(tf.square(hypothesis - Y))

train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
#복잡하게 한 것을 하나의 함수로 실행

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(100):
        print(step, sess.run(W))
        sess.run(train)
```

```python
import tensorflow as tf

X = [1, 2, 3]
Y = [1, 2, 3]

W = tf.Variable(5.)

hypothesis = X * W

gradient = tf.reduce_mean((W * X - Y) * X) * 2
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)

gvs = optimizer.compute_gradients(cost)
apply_gradients = optimizer.apply_gradients(gvs) #값을 apply

#자동으로 계산한 gradient와 미분의 수식을 넣은 gradient가 같은 값인지를 확인하겠다
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(101):
        gradient_val, w_val = sess.run([gradient, W])
        print(step, gradient_val, w_val) #gvs run
```
