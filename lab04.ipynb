```python
#lab04-01
import tensorflow as tf

x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
#x input이 3개, instance가 여러개 나왔을 경우 y의 값을 예측해보자.

x1 = tf.placeholder(tf.float32) #tf.float: data type
#placeholder: 실행시키는 단계에서 값을 넣어주고 싶을 때 사용
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name = 'weight1') #weight는 하나의 값
w2 = tf.Variable(tf.random_normal([1]), name = 'weight2')
w3 = tf.Variable(tf.random_normal([1]), name = 'weight3')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
hypothesis = x1*w1 + x2*w2 + x3*w3 + b

cost = tf.reduce_mean(tf.square(hypothesis - Y)) #hypothesis와 Y의 차의 제곱
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost) #cost가 가장 작게 optimize

sess = tf.Session()
sess.run(tf.global_variables_initializer()) #세션을 초기화 하는 순간 변수 W에 그 값이 지정
#변수들을 global_variables_initializer() 를 이용해서 초기화 한후, 초기화된 결과를 세션에 전달해 줘야 한다.

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                    feed_dict = {x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
                    #feed_dict에는 맨 위의 데이터 값을 주겠다. Y=우리가 목표로 하는 값
#feed_dict 변수를 이용해서 플레이스홀더 x에 위의 data를 피드하면, 
#세션에 의해서 그래프가 실행되면서 x는 feed_dict에 의해서 정해진 피드 데이터를 하나씩 읽어서 실행한다.

    if step % 10 == 0:
        print(step, "Cost:", cost_val, '\nPrediction\n', hy_val)
```

```python
import tensorflow as tf

x_data = [[73., 80., 75.], 
          [93., 88., 93.], 
          [89., 91., 90.], 
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.], [185.], [180.], [196.], [142.]]
#vector형식으로 저장

X = tf.placeholder(tf.float32, shape = [None, 3]) #x1,x2,x3의 세 개가 들어감, 각 element가 가지는 값은 3개
Y = tf.placeholder(tf.float32, shape = [None, 1]) #결과는 하나
#x_data는 이 데이터(인스턴스)가 몇개냐 하는건데 지금은 5개지만 그 이상일 수 있음
#따라서 n개를 가질 것이라서 None으로 적은 것임

W = tf.Variable(tf.random_normal([3,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.matmul(X, W) + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run(
    [cost, hypothesis, train], feed_dict = {X: x_data, Y: y_data})
    
    if step % 10 == 0:
        print(step, 'Cost: ', cost_val, '\nPrediction\n', hy_val)
```

```python
#lab04-02
#data가 너무 많아지는 경우는 텍스트 파일로 저장
import tensorflow as tf
import numpy as np
tf.set_random_seed(777)

xy = np.loadtxt('data-01-test-score.csv', delimiter = ',', dtype = np.float32)
#전체를 어레이로 읽어옴
x_data = xy[:, 0:-1] #slicing, [:,2]앞에 있는 값이 row, 두번째가 2번째 인덱스
#전체 row, 열은 0:-1까지(0부터 맨 마지막 인덱스 앞 인덱스까지)
y_data = xy[:, [-1]]
#전체 n개 다 가져온다, 맨 마지막 끝에 열만 가져오겠다.

print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data)
#shape이 맞는지를 확인 후 출력하기

X = tf.placeholder(tf.float32, shape=[None, 3]) 
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3,1]), name = 'weight')
#3개가 들어와서 1개의 y값이 나간다.[들어오는 값, 나가는 값]
b = tf.Variable(tf.random_normal([1]), name = 'bias')
#출력 y값은 하나[나가는 값]

hypothesis = tf.matmul(X, W) + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train],
        feed_dict = {X: x_data, Y: y_data})

#새 데이터 예측
print("Your score will be ", sess.run(hypothesis,
                        feed_dict = {X: [[100,70,101]]}))
print("Other scores will be ", sess.run(hypothesis,
                        feed_dict = {X: [[60,70,110], [90,100,80]]}))
```

```python
import tensorflow as tf

filename_queue = tf.train.string_input_producer(
['data-01-test-score.csv'], shuffle=False, name = 'filename_queue')
#file이 너무 큰 경우, file을 만들어서 queue

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)
#reader를 정의하는 부분

record_defaults = [[0.], [0.], [0.], [0.]] #각 필드의 데이터 타입
#디폴트 value는 빈 column.
xy = tf.decode_csv(value, record_defaults = record_defaults)
#csv로 decode하라

train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size = 10)
#pump로 데이터를 빨아올리듯이 가져온다. 한 번 펌프할 때마다 10개씩 가져와라

X = tf.placeholder(tf.float32, shape=[None, 3])#x value의 개수
Y = tf.placeholder(tf.float32, shape=[None, 1])#y value의 개수

W = tf.Variable(tf.random_normal([3,1]),name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')

hypothesis = tf.matmul(X, W) + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)

for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run(
    [cost, hypothesis, train], feed_dict = {X: x_batch, Y:y_batch})
    
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, '\nPrediction\n', hy_val)

coord.request_stop()
coord.join(threads)

# Ask my score
print("Your score will be ",
      sess.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))

print("Other scores will be ",
      sess.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
```
